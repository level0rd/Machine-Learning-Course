```python
import random as ran
import time
import math

def Okr(R, xc, yc, n):
    #xc,yc, R - Радиус и координаты центра 
    x = list(); y=list();
    ran.seed(time.time())
    
    for i in range(n):
        r = R * math.sqrt(ran.random())
        theta = ran.random() * 2 * math.pi
        x.append(r * math.cos(theta) + xc)
        y.append(r * math.sin(theta) + yc)
    
    return (x, y)

def Rect(R, xc, yc, n):
    x = list(); y=list();
    ran.seed(time.time())
    
    for i in range(n):
        x.append(ran.uniform(xc-R, xc+R))
        y.append(ran.uniform(yc-R, yc+R))
        
    return (x, y)
```

```python
res1 = Okr(15, 100, 100, 10)
res2 = Okr(50, 200, 150, 50)
res3 = Rect(20, 100, 180, 15)
res4 = Rect(10, 250, 50, 15)

import matplotlib.pyplot as plt;

plt.figure(figsize=(7,5), dpi=100)
plt.plot(res1[0], res1[1], 'o', color = "r", label='Точки в круг1');
plt.plot(res2[0], res2[1], 'o', color = "b", label='Точки в круг2');
plt.plot(res3[0], res3[1], 'o', color = "g", label='Точки в квадрат1');
plt.plot(res4[0], res4[1], 'o', color = "y", label='Точки в квадрат2');
plt.legend();
plt.show();
```

```python
def sinGen(n, noise=0):    
    x = list(); y=list()
    ran.seed(time.time())
    
    for i in range(n):
        x.append(i/10 + ran.random())
        y.append(math.sin(x[i])+ran.uniform(-noise, noise))                 
    return (x,y)
            
res1=sinGen(100)
res2=sinGen(100, 0.3)

plt.figure(figsize=(7,5), dpi=100)
plt.plot(res1[0], res1[1], 'o', color = "r", label='sin, noise = 0');
plt.plot(res2[0], res2[1], 'o', color = "b", label='sin, noise = 0.3');
plt.legend();
plt.show()
```
