# Сгенерировать 100 точек **НОРМАЛЬНО**  распределенных вокруг 5 со стандартным отклонением 2

```python
import numpy as np
import matplotlib.pyplot as plt

N = 100
mu, sigma = (5, 2)
numbers = np.random.normal(mu, sigma, N)
plt.plot(numbers, '.')
```

```python
from scipy.stats import norm
loc, scale = norm.fit(numbers)
loc, scale
```

# Сгенерировать 200 точек с **ПОКАЗАТЕЛЬНЫМ**(экспоненциальным) распределением с параметром λ=1

```python
import numpy as np
import matplotlib.pyplot as plt

N = 200
beta = 1
numbers = np.random.exponential(scale=1/beta, size=N)
plt.plot(numbers, '.')
```

```python
from scipy.stats import expon
loc, scale = expon.fit(numbers)
loc, scale
```

# Сгенерировать 500 точек **РАВНОМЕРНО** распределенных на интервале [-1, 3]

```python
import numpy as np
import matplotlib.pyplot as plt

N = 500
a, b = (-1, 3)
numbers = np.random.uniform(a, b, N)
plt.plot(numbers, '.')
```

```python
from scipy.stats import uniform
loc, scale = uniform.fit(numbers)
loc, scale
```

# Если p-value (pvalue) больше вашего уровня значимости (обычно 0.05), то нет оснований отклонить нулевую гипотезу о том, что наблюдаемые и ожидаемые частоты не отличаются значимо, и данные могут быть считаться порожденными нормальным распределением.
