    Для выявления скрытых зависимостей между этими наборами данных может использоваться Метод Наименьших Квадратов (далее МНК)
    Краткое описание метода:
    
    Пусть x — набор n неизвестных переменных (параметров), 
$$
f_i(x),i=1,...,m,m>n
$$

    - совокупность функций от этого набора переменных. Задача заключается в подборе таких значений x, чтобы 
    значения этих функций были максимально близки к некоторым значениям
    
$$
y_i.
$$

    По существу речь идет о «решении» переопределенной системы уравнений 
    
$$
f_i(x)=y_i,i=1,...,m
$$
    
    в указанном смысле максимальной 
    близости левой и правой частей системы. Суть МНК заключается в выборе в качестве
    «меры близости» суммы квадратов отклонений левых и правых частей 
    
$$
|f_i(x)-y_i|
$$  

    Таким образом, сущность МНК может быть выражена следующим образом:
        
$$
\sum_{i}e_i^2=\sum_{i}(y_i-f_i(x))^2\rightarrow min_x
$$

    В случае, если система уравнений имеет решение, то наименьшее значение суммы 
    квадратов будет равно нулю и могут быть найдены точные решения системы уравнений 
    аналитически или, например, различными численными методами оптимизации. 
    Если система переопределена, то есть, говоря нестрого, количество независимых уравнений 
    больше количества искомых переменных, то система не имеет точного решения и 
    метод наименьших квадратов позволяет найти некоторый «оптимальный» 
    вектор x в смысле максимальной близости векторов y и 
$$
f(x)
$$

    или максимальной близости вектора отклонений e к нулю (близость понимается в смысле
    евклидова расстояния). 

    Выражаем коэффициенты путем алгебраических преобразований. 



$$
\sum_{i=1}^n(y_i-(Ax_i+B))^2
$$

    Дифференцируем

\begin{equation*}
 \begin{cases}
   -2\sum_{i=1}^n(y_i-Ax_i^2-Bx_i)=0\\ 
   -2\sum_{i=1}^n(y_i-Ax_i-B)=0
 \end{cases}
\end{equation*}

\begin{equation*}
 \begin{cases}
   \sum_{i=1}^n(y_ix_i-Ax_i+B)=0\\ 
   \sum_{i=1}^n(y_i-Ax_i+B)=0
 \end{cases}
\end{equation*}

    Выражаем коэффициент В
                  
$$
\sum_{i=1}^nB=\sum_{i=1}^n(y_i-Ax_i)
$$

$$
nB=\sum_{i=1}^n(y_i-Ax_i)
$$

$$
B=\frac{\sum_{i=1}^ny_i+Ax_i}{n}=
\frac{\sum_{i=1}^ny_i-\sum_{i=1}^nAx_i}{n}
$$

    Подставляем его и приравниваем выражение к нулю

$$
\sum_{i=1}^n(y_ix_i-Ax_i^2-(\frac{\sum_{i=1}^ny_i-\sum_{i=1}^nAx_i}{n})x_i)=0
$$

$$
\sum_{i=1}^ny_ix_i-\frac{\sum_{i=1}^ny_i-\sum_{i=1}^nAx_i}{n}\sum_{i=1}^nx_i-A\sum_{i=1}^nx_i^2=0
$$

    Умножаем на n

$$
n\sum_{i=1}^ny_ix_i-(\sum_{i=1}^ny_i-\sum_{i=1}^nAx_in\sum_{i=1}^nx_i)-nA\sum_{i=1}^nx_i^2=0
$$

$$
n\sum_{i=1}^ny_ix_i-(\sum_{i=1}^ny_i\sum_{i=1}^nx_i-A(\sum_{i=1}^nx_i)^2)-nA\sum_{i=1}^nx_i^2=0
$$

$$
-nA\sum_{i=1}^nx_i^2+n\sum_{i=1}^ny_ix_i-\sum_{i=1}^nx_i\sum_{i=1}^ny_i+A(\sum_{i=1}^nx_i)^2=0
$$

$$
nA\sum_{i=1}^nx_i^2-n\sum_{i=1}^ny_ix_i+\sum_{i=1}^nx_i\sum_{i=1}^ny_i-A(\sum_{i=1}^nx_i)^2=0
$$

    Выражаем коэффициент А

$$
nA\sum_{i=1}^nx_i^2-A(\sum_{i=1}^nx_i)^2=n\sum_{i=1}^ny_ix_i-\sum_{i=1}^nx_i\sum_{i=1}^ny_i
$$

$$
A(n\sum_{i=1}^nx_i^2-(\sum_{i=1}^nx_i)^2)=n\sum_{i=1}^ny_ix_i-\sum_{i=1}^nx_i\sum_{i=1}^ny_i
$$

$$
A=\frac{n\sum_{i=1}^ny_ix_i-\sum_{i=1}^nx_i\sum_{i=1}^ny_i}{n\sum_{i=1}^nx_i^2-(\sum_{i=1}^nx_i)^2}
$$

    Подставляем его в B

$$
B=\frac{\sum_{i=1}^n(y_i-(\frac{n\sum_{i=1}^ny_ix_i-\sum_{i=1}^nx_i\sum_{i=1}^ny_i}{n\sum_{i=1}^nx_i^2-(\sum_{i=1}^nx_i)^2})x_i)}{n}
$$

$$
B=\frac{\sum_{i=1}^ny_i-\frac{n\sum_{i=1}^ny_ix_i-\sum_{i=1}^nx_i\sum_{i=1}^ny_i}{n\sum_{i=1}^nx_i^2-(\sum_{i=1}^nx_i)^2}\sum_{i=1}^nx_i}{n}
$$

#### Среднее значение

````python
def average(x):
    sum_num = 0
    for t in x:
        sum_num = sum_num + t           
    avg = sum_num / len(x)
    return avg

print("Среднее значение:", average(views))
print("Среднее значение(Numpy):", np.mean(views))
````

#### Дисперсия

````python
def dispersion(x, avg):
    variance = 0
    for i in x:
         variance += (avg - i) ** 2
    return variance / len(x)

print("Дисперсия:",dispersion(views, average(views)))
print("Дисперсия(Numpy):", np.var(views))
````

#### Коэффициент корреляции.

````python
import math

def coef_cor(x, y):
    s = 0
    for i,j in zip(x, y):
        s += (i-average(x))*(j-average(y))
    return 1/len(x)*(s/(math.sqrt(dispersion(x,average(x))*dispersion(y,average(y)))))

print("Коэффициент корреляции:", coef_cor(views, downloads))
print("Коэффициент корреляции(Numpy):", np.corrcoef(views, downloads))
````

#### Стандартное отклонение

````python
std = lambda x: math.sqrt(average(abs(x - average(x))**2))
    
print("Стандартное отклонение:", std(views))
print("Стандартное отклонение(Numpy):", np.std(views))
````

#### Максимальное и минимальное значение

````python
def mn(x):
    m = x[0]
    for i in range(len(x)):
        if x[i] < m:
            m = x[i]
    return m

def mx(x):
    m = x[0]
    for i in range(len(x)):
        if x[i] > m:
            m = x[i]
    return m

print("Минимальное значение просмотров:", mn(views))   
print("Минимальное значение просмотров:", mx(views)) 
print("Минимальное значение просмотров(Numpy):", np.amin(views))
print("Максимальное значение просмотров(Numpy):",np.amax(views))
````

### МНК

````python
def mnk(x,y):
        q=0; w=0;
        e=0;t=0;
        a=0;b=0;u=0;  
        if len(x)!=len(y):
            print("Размеры передаваемых списков должны быть одинаковыми")
            return [0,0] 
        else:
            N=len(y)
            try:
                for i in range(0,N):
                    q+=N*x[i]*y[i]
                    w+=x[i]
                    e+=y[i]
                r=q-w*e
                for i in range(0,N):
                    t=t+N*(x[i]**2)
                    u=u+x[i]
                u=(u**2)   
                for i in range(0,N):
                    a=r/(t-u)
                    b=1/N*(e-a*w)
                print("Коэффициент A:",a)
                print("Коэффициент В:",b)
                return [a,b]
            except:
                print("Значения хранимые в передаваемом списке должны быть числовыми или дробными")
                return [0,0]
            
Z=mnk(views,downloads)

x2=np.linspace(0,8000,100)
y2=Z[0]*x2+Z[1]

SomeSubPlot = plt.subplot(111)
SomeSubPlot.plot(x2,y2,'r', label=' Метод наименьших квадратов')
SomeSubPlot.plot(views,downloads,'*', label=' Рассеивание')
plt.ylabel('Просмотры')
plt.xlabel('Загрузки')
plt.title('МНК метод')
plt.savefig('img/linear_data.jpg')

chartBox = SomeSubPlot.get_position()
SomeSubPlot.set_position([chartBox.x0, chartBox.y0, chartBox.width, chartBox.height])
SomeSubPlot.legend(loc='upper center', bbox_to_anchor=(0.5, -0.2), shadow=True, ncol=1)
plt.show()
````

####  Линейная регрессия средствами Scikit-Learn

````python
from sklearn.linear_model import LinearRegression

def colvec(rowvec):
    v = np.asarray(rowvec)
    return v.reshape(v.size,1)

views = colvec(views)
downloads = colvec(downloads)

regr = LinearRegression()
regr.fit(views, downloads)
         
y2 = regr.predict(views)

SomeSubPlot = plt.subplot(111)
SomeSubPlot.plot(views,y2,'r', label=' Метод наименьших квадратов')
SomeSubPlot.plot(views,downloads,'*', label=' Рассеивание')

plt.ylabel('Просмотры')
plt.xlabel('Загрузки')
plt.title('МНК метод - встроенные функции')

print("Коэффициент A:", regr.coef_)
print("Коэффициент В:",regr.intercept_)

chartBox = SomeSubPlot.get_position()
SomeSubPlot.set_position([chartBox.x0, chartBox.y0, chartBox.width, chartBox.height])
SomeSubPlot.legend(loc='upper center', bbox_to_anchor=(0.5, -0.2), shadow=True, ncol=1)
plt.show()
````
