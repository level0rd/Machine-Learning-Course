## Просто цикл ("классический" подход)

````python
%%time

counts = {}

for line in ALICE_LINES:
    words = tokenizer.tokenize(line)

    for word in words:
        word = word.lower()
        counts[word] = counts.get(word, 0) + 1

print(counts)
print(sum(counts.values()))
list(
    reversed(
        sorted((n, word) for word, n in counts.items())
    )
)[:10]
````

## Идиоматический Питон


````python
words = [w for line in ALICE_LINES for w in tokenizer.tokenize(line)]
print(words[:50])

lower_words = [w.lower() for w in words]
print(lower_words[:50])

print(len(lower_words))
````

````python
%%time
from collections import Counter

counts = Counter(lower_words)
print(sum(counts.values()))

counts.most_common(10)
````

````python
%time Counter(w.lower() for line in ALICE_LINES for w in tokenizer.tokenize(line)).most_common(10)
````

````python
%%time
import operator
from functools import reduce

words = [[w for w in tokenizer.tokenize(line)] for line in ALICE_LINES]
print(words[:4])

lower_words = [[w.lower() for w in line] for line in words]
print(lower_words[:4])

counters = [Counter(line) for line in lower_words]
print(counters)

reduce(operator.add, counters, Counter())\
    .most_common(10)
````

## Функциональный подход

````python
%%time
from itertools import chain

word_lists = list(map(tokenizer.tokenize, ALICE_LINES))
print('word_lists: ', word_lists[:3])

words = list(chain(*word_lists))
print('words: ', words[:20])

lower_words = list(map(str.lower, words))
print('lower_words: ', lower_words[:10])

Counter(lower_words).most_common(10)
````

````python
%%time

word_lists = list(map(tokenizer.tokenize, ALICE_LINES))
print(word_lists[:4])

lower_words = list(map(
    lambda line_words: [w.lower() for w in line_words],
    word_lists
))
print(lower_words[:4])

counters = map(Counter, lower_words)
reduce(operator.add, counters, Counter()).most_common(10)
````

## Удобные свойства map и reduce:

Если функция чистая, то мап можно вызывать параллельно

[Python Pool](https://docs-python.ru/standart-library/paket-multiprocessing-python/klass-pool-modulja-multiprocessing/)

````python
from multiprocessing import Pool

print(list(map(str, [1, 2, 3, 4, 5, 6])))

with Pool() as pool:
  print(pool.map(str, [1, 2, 3, 4, 5, 6]))
````

Если reduce действительно работает как reduce, то его можно вызывать от результата других вызовов reduce

````python
from functools import reduce

def add(s, x):
  return s + x

print(reduce(add, range(10)))

print(reduce(add, 
       [
        reduce(add, range(3)), 
        reduce(add, range(3,6)),
        reduce(add, range(6,10))
       ]
))
````

````python
def count(s, x):
  return s + 1

print(reduce(count, range(10)))

print(reduce(count, 
       [
        reduce(count, range(3), 0), 
        reduce(count, range(3,6), 0),
        reduce(count, range(6,10), 0)
       ]
      , 0)
)
````

## Локальный MapReduce
https://towardsdatascience.com/a-beginners-introduction-into-mapreduce-2c912bb5e6ac

````python
from collections import Counter
from itertools import chain
from multiprocessing import Pool
from functools import reduce
import operator
from nltk.tokenize import RegexpTokenizer

tokenizer = RegexpTokenizer(r'\w+')
````

````python
from typing import List, Iterable

def wordcounter(s: str) -> Counter:
  return Counter(map(str.lower, tokenizer.tokenize(s)))

wordcounter("Hello world!")
````

````python
def reduce_counters(counters: Iterable[Counter]) -> Counter:
  return reduce(operator.add, counters, Counter())
  
%time reduce_counters(map(wordcounter, ALICE_LINES)).most_common(10)
````

````python
from multiprocessing import Pool

with Pool() as pool:
  %time reduce_counters(pool.map(wordcounter, ALICE_LINES)).most_common(10)
````

````python
def word_counters(s: str) -> List[Counter]:
  return [Counter([w.lower()]) for w in tokenizer.tokenize(s)]

word_counters("Hello world!")
````

````python
def wordcount_local(texts: Iterable[str]) -> Counter:
    return reduce_counters(
        map(reduce_counters,
            map(word_counters,
                texts)
        )
    )
````

````python
%time wordcount_local(ALICE_LINES).most_common(10)
````

````python
def wordcount_local_mp(texts: Iterable[str]) -> Counter:
  with Pool() as pool:
    return reduce_counters(
      pool.map(reduce_counters,
        pool.map(word_counters,
          ALICE_LINES)
      )
  )
````

# Простой "распределенный" MapReduce

https://ru.wikipedia.org/wiki/MapReduce

mapreduce.png

Принцип обработки данных:
```
[записи]
-> Map ->
[(k1, v11), (k2, v21), (k1, v12) ...]
-> Shuffle ->
{k1: [v11, v12, ...]), k2: [v21, ...], ...}
-> Reduce ->
{k1: V1, k2: V2, ...}
```

Действие функций mapper и reducer, решающих задачу:

`запись -> mapper ---(output)---> [(k1, v1), (k2, v2), (k1, v3) ...]`

`[v1, v3, ...] -> reducer ---(output)---> V`

## Map

````python
from typing import List, Tuple

def count_words_mapper(line: str) -> List[Tuple[str, int]]:
  return [(w, 1) for w in tokenizer.tokenize(line)]
````

````python
count_words_mapper("hello world")
````

````python
list(map(count_words_mapper,
         ["hello world", "hello me"]
         ))
````

````python
from typing import Callable

def map_layer(mapper: Callable) -> Callable:
  def inner(records: Iterable) -> List[Tuple]:
    return [(k, v) for r in records for k, v in mapper(r)]
  return inner

word_counters = map_layer(count_words_mapper)
````

````python
word_counters( ["hello world", "hello me"])
````

````python
%time list(word_counters(ALICE_LINES))[:20]
````

## Merge by key (Shuffle)


````python
from typing import Dict, Any
from collections import defaultdict

def merge_layer(pairs: List[Tuple]) -> Dict[Any, List]:
  result = defaultdict(list)
  for k, v in pairs:
    result[k].append(v)
  return dict(result)

merge_layer([('a', 1), ('b', 2), ('c', 3), ('a', 4), ('a', 5), ('x', 6), ('y', 7)])
````

````python
from itertools import islice
merge_layer(word_counters(islice(ALICE_LINES, 20)))
````

## Reduce

`[v1, v3, ...] -> reducer -> V`

`{k1: [v11, v12, ...]), k2: [v21, ...], ...} -> Reduce -> {k1: V1, k2: V2, ...}`

````python
def reduce_layer(reducer: Callable) -> Callable:
  def inner(merged: Dict[Any, List]) -> Dict:
    return {k: reducer(v) for k, v in merged.items()}
  return inner

counter_adder = reduce_layer(sum)

alice_wordcounts = counter_adder(merge_layer(word_counters(ALICE_LINES)))

print(alice_wordcounts)
````

````python
sorted(alice_wordcounts.items(), key=lambda pair: -pair[1])[:20]
````

## Собираем

````python
import pandas as pd


def simple_job(mapper: Callable, reducer:Callable) -> Callable:
  m = map_layer(mapper)
  r = reduce_layer(reducer)
  def inner(records: List) -> Dict:
    return dict(r(merge_layer(m(records))))
  return inner


wordcounter = simple_job(
    lambda f: ((token, 1) for token in tokenizer.tokenize(f)),
    sum)
gutenberg_raw = [gutenberg.raw(f) for f in gutenberg.fileids()]


%time result = wordcounter(gutenberg_raw)

pass
sorted(result.items(), key=lambda pair: -pair[1])[:20]
````

Модель распределённых вычислений **MapReduce**, компонент Apache Hadoop для обработки данных, проводит вычисления в два этапа:

1. **`Map`**, когда главный узел кластера (master) распределяет задачи по рабочим узлам (node).

    <img height=400 src="https://habrastorage.org/r/w1560/getpro/habr/post_images/25b/ceb/dec/25bcebdec55f86c7c49a9bfe862ce21d.jpg"/>

    Принимает на вход список значений и некую функцию, которую затем применяет к каждому элементу списка и возвращает новый список. Выполняется параллельно и локально над каждым блоком данных. Вместо того, чтобы доставлять терабайты данных к программе, небольшая, определённая пользователем программа копируется на сервера с данными и делает с ними всё, что не требует перемешивания и перемещения данных (_shuffle_).

2. **`Shuffle`**, когда рабочие узлы перераспределяют данные на основе ключей, ранее созданных функцией _Map_, таким образом, чтобы все данные одного ключа лежали на одном рабочем узле.

2. **`Reduce`**, когда данные сворачиваются и передаются обратно на главный узел, формируя окончательный результат вычислений.

![](https://www.bigdataschool.ru/wp-content/uploads/2019/10/MapReduce3.png)

Стандартный MapReduce спроектирован так, что все результаты — как конечные, так и промежуточные — записываются на диск. Поэтому все операции проходят по циклу чтение-запись с жесткого диска.
