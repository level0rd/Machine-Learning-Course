```python
print(__doc__)

# Источник кода: Гаэль Вароку
# Изменено для документации Жаком Гроблером
# Лицензия: пункт 3 BSD

import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LogisticRegression
from sklearn.datasets import load_iris

# импортировать данные для игры
iris = load_iris()
X = iris.data[:, :2]  # мы берем только первые две функции.
Y = iris.target

# Создать экземпляр классификатора логистической регрессии и подогнать под данные.
logreg = LogisticRegression()
logreg.fit(X, Y)


#-------------------------------------------------------------------------------
# Постройте границу решения. Для этого мы присвоим каждому цвету
# точка в сетке [x_min, x_max] x [y_min, y_max].
x_min, x_max = X[:, 0].min() - .5, X[:, 0].max() + .5
y_min, y_max = X[:, 1].min() - .5, X[:, 1].max() + .5
h = .02  # размер шага в сетке
xx, yy = np.meshgrid(np.arange(x_min, x_max, h), np.arange(y_min, y_max, h))
Z = logreg.predict(np.c_[xx.ravel(), yy.ravel()])

# Поместите результат в цветной график
Z = Z.reshape(xx.shape)
plt.figure(1, figsize=(15,8))
plt.pcolormesh(xx, yy, Z, cmap='Spectral') #plt.cm.Paired

# Постройте также тренировочные точки
plt.scatter(X[:, 0], X[:, 1], s=160, c=Y, edgecolors='k', cmap='Spectral', linewidths=2)
plt.xlabel('Sepal length', fontsize=24)
plt.ylabel('Sepal width', fontsize=24)

plt.xlim(xx.min(), xx.max())
plt.ylim(yy.min(), yy.max())
plt.xticks(())
plt.yticks(())

p = plt.imshow(Z, cmap='Spectral')
crb=plt.colorbar(p)
crb.ax.tick_params(labelsize=24)
plt.show()
```
