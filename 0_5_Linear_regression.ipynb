```python
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression
import pandas as pd


def colvec(rowvec):
    v = np.asarray(rowvec)
    return v.reshape(v.size,1)

data = pd.read_csv('AllD/cite.csv')

vievs = np.array(data['vievs'])

downloads = np.array(data['downloads'])


vievs = colvec(vievs)
downloads = colvec(downloads)



regr = LinearRegression()
regr.fit(vievs, downloads)

A=regr.coef_
B=regr.intercept_

print(A)
print(B)

yy = regr.predict(vievs)

SomeSubPlot = plt.subplot(111)
SomeSubPlot.plot(vievs,yy,'r', label=' Метод наименьших квадратов')
SomeSubPlot.plot(vievs,downloads,'*', label=' Рассеивание')

plt.ylabel(u'Просмотры')
plt.xlabel(u'Загрузки')
plt.title(u'МНК метод - встроенные функции')


chartBox = SomeSubPlot.get_position()
SomeSubPlot.set_position([chartBox.x0, chartBox.y0, chartBox.width, chartBox.height])
SomeSubPlot.legend(loc='upper center', bbox_to_anchor=(0.5, -0.2), shadow=True, ncol=1)
plt.show()
```

```python

import seaborn as sns; sns.set()


plt.figure(figsize=(5, 5), dpi=200)
plt.plot(vievs,yy,'r', label=' Метод наименьших квадратов')
plt.scatter(vievs, downloads, edgecolor="red", s=50,cmap='coolwarm', label="Рассеивание");

plt.ylabel("Загрузки")
plt.xlabel("Просмотры")
plt.title("Зависимость зарегистрировавшихся на сайте от количества посещений за 10 дней")
plt.legend()
plt.show()
```
