#### Отчет по теме: "Изучение функций активации нейронных сетей"
    В нейронных сетях функция активации нейрона определяет
    выходной сигнал, который определяется входным сигналом или набором 
    входных сигналов. 
    
    Linear
    Функция прямой линии, где активация пропорциональна входу (который является 
    взвешенной суммой от нейрона).
$$
R(z,m) = {z * m}
$$
    ELU
    Экспоненциальная линейная функция. Эта функция, которая, как правило, 
    быстрее сводит затраты к нулю и дает более точные результаты.
    
\begin{cases}x     
  & \text{if $x>0$,}\\
  α(exp(x)-1)) & \text{if $x<=0$.}
\end{cases}

![1](Grapth\Elu.png)
    
    ReLU
    Реализует простой пороговый переход в нуле.
$$
f(x) = max(0,x)
$$   

![1](Grapth\Relu.png)

    Tanh
    Гиперболический тангенся, является скорректированной формой сигмоида.
$$
tanh(z) = \frac{e^z-e^{-z}}{e^z+e^{-z}}
$$   

![1](Grapth\Tanh.png)

    Sigmoid
    Sigmoid принимает значение на входе и выводит другое 
    значение в диапазоне от 0 до 1. С ним легко работать, и он 
    обладает всеми необходимыми свойствами функций активации: он 
    нелинейный, непрерывно дифференцируемый, монотонный и 
    имеет фиксированный диапазон выходных данных.
$$
R(z) = \frac{1}{1+e^{-z}}
$$

![1](Grapth\Sigmoid.png)

    Softmax
    Функция преобразует вектор z  размерности K в вектор R той же 
    размерности, где каждая координата R полученного вектора 
    представлена вещественным числом в интервале [0,1] и сумма 
    координат равна 1
$$
R(z)_i = \frac{e^{z_i}}{\sum_{j=1}^Ke^{z_j}}
$$   



    Softplus
$$
f(x) = ln(1 + e^x)
$$
