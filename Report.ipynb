# Функции, разработанные на основе вышеописанного


````python
import random as ran
import time
import math as math
import numpy as np

#Функции для генерации одной финугры


def Okr(R=15, xc=50, yc=50, n=100):
    # Выдает данные в виде круга
    # Параметры:
    # R - размер
    # xc - координата центра по х
    # уc - координата центра по у
    # n - количетсво эдементов
    
    # Возвращает массив, содержищий массивы с координатами
    
    
    ran.seed(time.time())
    data = list();    
    for i in range(n):
        r = R * math.sqrt(ran.random())
        theta = ran.random() * 2 * math.pi
        data.append([r * math.cos(theta) + xc,r * math.sin(theta) + yc])         
    return data


def Rect(R=15, xc=50, yc=50, n=100):
    # Выдает данные в виде квадрата
    # Параметры:
    # R - размер
    # xc - координата центра по х
    # уc - координата центра по у
    # n - количетсво эдементов
    
    # Возвращает массив, содержищий массивы с координатами
    
    
    ran.seed(time.time())
    data = list();    
    for i in range(n):
        data.append([ran.uniform(xc-R, xc+R),ran.uniform(yc-R, yc+R)])        
    return data


def RAr(R=15, xc=50, yc=50, n=100):
    # Выдает данные в виде случайной фигуры
    # Параметры:
    # R - размер
    # xc - координата центра по х
    # уc - координата центра по у
    # n - количетсво эдементов
    
    # Возвращает массив, содержищий массивы с координатами
    
    ran.seed(time.time())
    data = list();    
    for i in range(n):
        data.append([ran.gauss(xc,0.5*R), ran.gauss(yc,0.5*R)])
    return data


def RAr1(R=15,c=50, n=100, nf=2):
    # Выдает данные в виде случайной фигуры, можно использовать несколько признаков
    # Параметры:
    # R - размер
    # c - координата центра
    # n - количетсво эдементов
    # nf - количество возвращаемых признаков
    
    # Возвращает массив, содержищий массивы с координатами
    
    
    ran.seed(time.time())
    data = list();
    for i in range(n):
        feats=[]
        for j in range(nf):
            feats.append(ran.gauss(c,0.5*R));
        data.append(feats)
    return data

# -----------------------------------------------------
# Функция для генерации множества классов

def generate(mass, dist=1.0):
    # Данная функция генерирует множество фигур
    # mass - массив из фигур
    # dist - расстояние
    #
    # элемент массива фигур имеет вид ['фигура', кол-во точек, размер]
    # Виды фигур; 
    # - circle - круги,
    # - rect - квадраты,
    # - rrr - произвольная.
    '''
    Пример массива 
        a1=(['circle', 70, 15], 
            ['circle', 62, 15],
            ['rect', 81, 15],
            ['circle',63, 7],
            ['blob', 50, 15])
    '''
    
    # Возвращает два массива. Первый содержит координаты, второй содержит индексы классов  
    
    
    maxX=dist*100;
    maxY=dist*100;    
    x = list(); # Содержит координаты
    y = list(); # содержит класс
    i = -1;
    for f,n,r in (mass):
        i+=1
        centerX, centerY = ran.random()*maxX, ran.random()*maxY;
        if (f=='circle'):            
            #x = x + Okr(r, centerX, centerY, n)
        if (f=='rect'):            
            x += Rect(r, centerX, centerY, n)
        if (f=='blob'):
            x += RAr(r, centerX, centerY, n)
        y += ([i] * n);
        
    return np.array(x)/dist,y;


def generate_blobs(n_classes=3, n_features=2, n_elements=100, dist=10): 
    '''
    Функция генерирует набор классов с различным количеством признаков
    Параметры
    n_classes - Количетсво классов
    n_features - количество признаков
    n_elements - уоличетсво элементов в каждом классе
    dist - расстояние между классами
    
    Возвращает массив признаков и массив принадлежности к классу
    '''
    
    
    ran.seed(time.time())
    data = list();
    y = list();
    centers = [0]*n_features
    
    for i in range(n_classes):
        for j in range(n_features):            
            centers[j] = ran.random()*dist;            
        for k in range(n_elements):
            feats=[]
            for j in range(n_features):              
                #feats.append(ran.gauss(centers[j],0.5));
            data.append(feats)
            y += ([i])
    return np.array(data),y

#-----------------------------------------------------

def GenByPars(Params,Funct,Noise=0,Start=0,End=10,PointsCount=50):
    # Данная функция генерирует точки из математической функции c шумом
    '''
    Парметры:
    Params - Строка содержит в себе параметры математической функции. Параметры разделяются ';'. 
             Например 'a=3;b=5;'
    Funct - Строка содержит в себе математическую функцию. Значение встроенного параметра 'X' берется автоматически.
            Например, 'a*np.cos(X+b)',
                      'a*X+b',
                      'X**2+a'
    Noise - Шум
    Start - Начальная точка
    End - Конечная точка
    PointsCount - Количество точек
    '''
    
    
    if (isinstance(Params, str)==False) | (isinstance(Funct, str)==False):
        print('Params и Funct должны быть строками')
        return 0
    
    StringEx = '';
    G=Params.find(";")
    while (G!=-1):
        StringEx = Params[: Params.find(";")]
        exec(StringEx)
        Params = Params[ Params.find(";") + 1 : ]
        G = Params.find(";")
         
    
    X = np.linspace(Start, End, PointsCount, True)
    Y = eval(Funct)    
    
    for i in range(0,len(Y)):             
        Y[i] = Y[i] + np.random.uniform(-Noise,Noise)
        X[i] = X[i] + np.random.uniform(-Noise,Noise)
            
    Result = np.array([X, Y])
    return Result
````
